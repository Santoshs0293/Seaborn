```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

```python
sns.set(style="whitegrid")
pokemon=pd.read_csv('pokemon.csv')
```

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Attack", data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Speed",palette='rocket', data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Speed",palette='vlag', data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon, color="green")
plt.show()
```

```python
loan=pd.read_csv('loan.csv')
loan.head()
```

```python
sns.scatterplot(x="ApplicantIncome",y="LoanAmount", data=loan)
plt.show()
```

```python
sns.scatterplot(x="ApplicantIncome",y="LoanAmount", data=loan, hue="CoapplicantIncome")
plt.show()
```

```python
sns.distplot(loan['LoanAmount'])
plt.show()
```

```python
sns.distplot(loan['LoanAmount'],color="green",bins=5,kde=False)
plt.show()
```

```python
sns.distplot(loan['LoanAmount'],color="green",vertical=True)
plt.show()
```

```python
sns.jointplot(x='LoanAmount',y="ApplicantIncome",data=loan)
plt.show()
```

```python
sns.jointplot(x='LoanAmount',y="ApplicantIncome",data=loan,color="red",kind="reg")
plt.show()
```

```python
sns.boxplot(x="LoanAmount",y="Education",data=loan)
plt.show()
```

```python
sns.boxplot(x="LoanAmount",y="Education",data=loan,palette="Set1")
plt.show()
```
